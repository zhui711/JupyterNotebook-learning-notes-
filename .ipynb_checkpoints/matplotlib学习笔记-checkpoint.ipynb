```python
#part2：多图形绘制
#铁律：画一幅图要在一个代码块内完成，不能分块
```

```python
#准备数据
x=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[0,0,0,0,0]
y3=[-1,-2,-3,-4,-5]
```

```python
#以matlab方式
Fig1=plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
#集成在一个图内
```

```python
#绘制多个子图
Fig2=plt.figure()
plt.subplot(3,1,1),plt.plot(x,y1)
plt.subplot(3,1,2),plt.plot(x,y2)     #(3,1,2)表示子图共3行1列，此子图是第2个
plt.subplot(3,1,3),plt.plot(x,y3)
```

```python
#统计图
import numpy as np
data=np.random.randn(10000)  #创建10000个标准正态分布样本
```

```python
Fig1=plt.figure()
plt.hist(data)   #hist就是直方图
```

```python
#bins=xx  ————表示区间划分数量
Fig1=plt.figure()
plt.hist(data,bins=30)    #原来默认10条柱子现在改成了30
```

```python
#alpha=xx   ————透明度，默认是1
Fig1=plt.figure()
plt.hist(data,alpha=0.5)
```

```python
#part4：图窗属性
```

```python
#设置坐标轴上下限：lim法  or  axis法
```

```python
#准备数据
x=[1,2,3,4,5]
y=[1,8,27,64,125]
```

```python
#lim法
Fig1=plt.figure()
plt.plot(x,y)
plt.xlim(1,5)      #x轴的范围是1-5
plt.ylim(1,125)    #y轴的范围是1-125
```

```python
#axis法 ————一行代码解决
Fig2=plt.figure()
plt.plot(x,y)
plt.axis([1,5,1,125])
```

```python
#标题与轴名称:标题就是plt.titel；轴名称就是：plt.x/ylabel
Fig1=plt.figure()
plt.plot(x,y)
plt.title('This is the title')
plt.xlabel('This is the xlabel')
plt.ylabel('This is the ylabel')
```

```python
#图例：plt.legend()即可
Fig1=plt.figure()
plt.plot(x,y1,label='y=x')
plt.plot(x,y2,label='y=0')
plt.plot(x,y3,label='y=-x')
plt.legend()
#若不想展示某些线条的图例只需去除该函数中的label关键字即可
#想调整这个图例的位置取改它的关键字即可，还能去边框
```

```python
#加网格：只需要在最后加个plt.grid()
Fig1=plt.figure()
plt.plot(x,y1,label='y=x')
plt.plot(x,y2,label='y=0')
plt.plot(x,y3,label='y=-x')
plt.legend()
plt.grid()
#grid还有color和linestyle两个关键字
```
