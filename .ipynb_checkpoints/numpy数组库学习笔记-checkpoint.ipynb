```python
arr=np.random.normal(500,70,10000)
print(np.all(arr>250))   #åˆ¤æ–­æ˜¯å¦æ‰€æœ‰çš„æ•°æ®éƒ½å¤§äº250
```

```python
#ä»¥å¸ƒå°”å‹æ•°ç»„ä½œä¸ºindexï¼ˆæ©ç ï¼‰ç­›é€‰ç¬¦åˆæ¡ä»¶çš„å…ƒç´ â€”â€”â€”â€”â€”â€”æœ€é‡è¦çš„åŠŸèƒ½
arr=np.arange(1,13).reshape(3,4)
print(arr)
print(arr>4)  #arr>4å·²ç»æ˜¯ä¸€ä¸ªæ–°çš„å¸ƒå°”å‹æ•°ç»„äº†
print(arr[arr>4])
#å¸ƒå°”æ•°ç»„çš„æ¯ä¸ªå¸ƒå°”å€¼ï¼ˆTrue/Falseï¼‰ä¸åŸæ•°ç»„çš„æ¯ä¸ªå…ƒç´ ä¸€ä¸€å¯¹åº”
#æ³¨æ„ï¼šå¯¹çŸ©é˜µè¿›è¡Œæ©ç æ“ä½œåé€€åŒ–ä¸ºäº†å‘é‡
```
